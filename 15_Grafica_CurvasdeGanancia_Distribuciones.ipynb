```python
# Crear un nuevo DataFrame para almacenar las ganancias maximas: promedio(Gan_max, Gan_lag(-250), Gan_lead(+250))
df_GanMax = resumen[resumen.columns[:2]].copy()
df_GanMax['Gan_Baseline'] = pd.Series(dtype=int)
df_GanMax['Gan_rank0fix'] = pd.Series(dtype=int)
df_GanMax['Gan_ranksimple'] = pd.Series(dtype=int)
df_GanMax['Gan_scale'] = pd.Series(dtype=float)
```

```python
#Calcula Ganancia Maxima para: BASELINE
#Tomamos el df con todas las ganancias para cada cant. envios:
dataset = df_baseline.iloc[:,1:]

# Encontrar el índice del valor máximo en cada columna
max_indices = dataset.idxmax()

# Calcular el promedio de la ganancia máxima y sus vecinos inmediatos para cada columna
i=0
for columna in dataset.columns:
    indice_max = max_indices[columna]
    
    # Asegúrate de que los índices vecinos estén dentro de los límites del DataFrame
    indice_lag = max(indice_max - 1, 0)
    indice_lead = min(indice_max + 1, len(dataset) - 1)
    
    # Calcula el promedio de las ganancias máximas y vecinas
    ganmax_avg = dataset.iloc[[indice_lag, indice_max, indice_lead]][columna].mean().astype(int)
    
    # Almacena el resultado en el nuevo DataFrame
    df_GanMax.loc[i,'Gan_Baseline'] = ganmax_avg
    i= i+1

#borramos lo grabado:    
del dataset
del max_indices
```
